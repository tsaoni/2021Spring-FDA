````python
# check which column contains null value
df.isna().any()[lambda x : x]
````

````python
# 資料視覺化
# plot pclass
pclass = []
x = np.arange(1, 4)
pclass = np.array([[0, 0], [0, 0], [0, 0]])
for i in range(0, len(df)):
  pclass[df.iloc[i]["Pclass"] - 1][df.iloc[i]["Survived"]] += 1

width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, pclass[:, 0], width, label='unsurvived')
rects2 = ax.bar(x + width/2, pclass[:, 1], width, label='survived')

ax.set_ylabel('population')
ax.set_title('Pclass')
ax.set_xticks(x)
ax.legend()

def autolabel(rects):
  for r in rects:
    height = r.get_height()
    ax.annotate('{}'.format(height),
          xy=(r.get_x() + r.get_width()/2, height),
          xytext=(0, 3),
          textcoords="offset points",
          ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
# plt.bar(x, pclass[:, 0])
plt.show()
````

#### 觀察:

* 人數絕大多數集中在class 3
* class 3中又以unsurvived人數最多

````python
# plot sex
sex = []
labels = ["male", "female"]
x = np.arange(1, 3)
sex = np.array([[0, 0], [0, 0]])
for i in range(0, len(df)):
  sex[0 if df.iloc[i]["Sex"] == "male" else 1][df.iloc[i]["Survived"]] += 1

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, sex[:, 0], width, label='unsurvived')
rects2 = ax.bar(x + width/2, sex[:, 1], width, label='survived')

ax.set_ylabel('population')
ax.set_title('Sex')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
````

````python
# 資料視覺化
# plot age
age_range = 5
x = np.arange(0, int(100/age_range))
age = np.empty((0, 2), int)
labels = np.array([])
for i in range(0, int(100/age_range)):
  age = np.append(age, np.array([[0, 0]]), axis=0)
  labels = np.append(labels, str(i*age_range) + "-" + str((i+1)*age_range))
for i in range(0, len(df)):
  if not pd.isnull(df.iloc[i]["Age"]):
    ind = int(df.iloc[i]["Age"] / age_range)
    age[ind][df.iloc[i]["Survived"]] += 1

fig, ax = plt.subplots(figsize=(12, 4.8))
rects1 = ax.bar(x - width/2, age[:, 0], width, label='unsurvived')
rects2 = ax.bar(x + width/2, age[:, 1], width, label='survived')

ax.set_ylabel('population')
ax.set_title('Age')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
````

````python
# classify and calculate kinds of tickets
cag = [""]
name = np.array([])
for i in range(0, len(df)):
  s1 = df.iloc[i]["Ticket"].split(" ", 1)
  name = np.append(name, None)
  name[-1] = s1

def inner_tic(cag, nm):
  for c in cag:
    if len(nm) == 1 or c == nm[0]:
      return
  cag = cag.append(nm[0])

for i in range(0, len(name)):
  inner_tic(cag, name[i]) 
print(len(cag))
````

````python
# plot ticket
labels = np.copy(cag)
tcnum = len(labels)
x = np.arange(0, tcnum)
tic = np.empty((0, 2), int)
for i in range(0, tcnum):
  tic = np.append(tic, np.array([[0, 0]]), axis=0)

for i in range(0, len(df)):
  s = df.iloc[i]["Ticket"].split(" ", 1)
  for k in range(0, tcnum):
    if len(s) == 1:
      tic[0][df.iloc[i]["Survived"]] += 1
      break
    elif labels[k] == s[0]:
      tic[k][df.iloc[i]["Survived"]] += 1
      break

fig, ax = plt.subplots(figsize=(48, 10))
rects1 = ax.bar(x - width/2, tic[:, 0], width, label='unsurvived')
rects2 = ax.bar(x + width/2, tic[:, 1], width, label='survived')

ax.set_ylabel('population')
ax.set_title('Ticket')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
````

````python
# find all value of embarked
cag = []

def inner(cag, cpr):
  for c in cag:
    if c == cpr:
      return
  cag = cag.append(cpr)
  return

for i in range(0, len(df)):
  if not pd.isnull(df.iloc[i]["Embarked"]):
    inner(cag, df.iloc[i]["Embarked"])
  
print(cag)
````

找出所有embarked的種類
共有S, C, Q三種

````python
# plot embarked
labels = ['S', 'C', 'Q']
x = np.arange(0, 3)
em = np.array([[0, 0], [0, 0], [0, 0]])

def ret_em(em):
  return {
  'S': 0,
  'C': 1,
  'Q': 2
  }.get(em, -1)
for i in range(0, len(df)):
  if not pd.isnull(df.iloc[i]["Embarked"]):
    em[ret_em(df.iloc[i]["Embarked"])][df.iloc[i]["Survived"]] += 1

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, em[:, 0], width, label='unsurvived')
rects2 = ax.bar(x + width/2, em[:, 1], width, label='survived')

ax.set_ylabel('population')
ax.set_title('Embarked')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
````

#### 觀察:

#### 處理缺失值方法
```
缺失值填補我是自己寫fill_null的function，依據各項feature種類量比率，用random的方式做填補

例如說，假設某欄位features A1, A2, A3, ..., An
分別有X1, X2, ..., Xn個人數，人數總量X
要去填補某個空值
我就先random出一個介於0到X之間的數p
再用迴圈判斷p落於0-X1, X1-X2, ..., Xn-1-Xn
哪個區間內，若落於Xi-1-Xi:
(i) 若是類別型資料，即用種類Ai填補空值
(ii) 若是數值型資料，則用i乘以統計時用的區間來填補
```

* Age跟Embarked原本是字串的資訊，在傳入fill_null前有先把他做字串轉數值
* Age跟Fare在統計中分別是以5跟50做一個區間，在填補這兩個欄位的空值時，用選取的種類i分別乘以5跟50作為填補值

## 模型訓練

請在此區塊後使用**一個或多個區塊**進行模型訓練，**需要**包含以下內容：

- **選擇模型**
    - 設定模型**超參數**
    - **控制隨機亂數**
- 計算**準確度（Accuracy）**
    - 使用 **5-fold cross validation**，輸出**平均準確度**
    - 最後再使用**所有**資料進行訓練並輸出**準確度**
    - 準確度必須**超越**課堂使用的範例
    
如果有進行多於一個以上的實驗，**可以保留所有**的實驗結果，但是執行時間將會納入**全部**實驗。

````python
# 模型訓練 (hw original ver) (decision tree)

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型
from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model = DecisionTreeClassifier(random_state=1012) # 創造決策樹模型
    model.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_pred_y = model.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
````

#### decision tree

##### 演算法解釋
決策樹是一個預測模型；他代表的是對象屬性與對象值之間的一種映射關係。樹中每個節點表示某個對象，而每個分叉路徑則代表某個可能的屬性值，而每個葉節點則對應從根節點到該葉節點所經歷的路徑所表示的對象的值

````python
# 模型訓練 (Stochastic Gradient Descent)

from sklearn.model_selection import KFold
from sklearn.linear_model import SGDClassifier    
from sklearn.metrics import accuracy_score  

kf = KFold(n_splits=5,             
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x) 

train_acc_list = []
valid_acc_list = []

for train_index, valid_index in kf.split(train_x):
    train_x_split = train_x.iloc[train_index]
    train_y_split = train_y.iloc[train_index]
    valid_x_split = train_x.iloc[valid_index]
    valid_y_split = train_y.iloc[valid_index]
    
    model = SGDClassifier(loss="hinge", penalty="l2", max_iter=50)
    model.fit(train_x_split, train_y_split)
    
    train_pred_y = model.predict(train_x_split)
    train_acc = accuracy_score(train_y_split, train_pred_y)
    valid_pred_y = model.predict(valid_x_split)
    valid_acc = accuracy_score(valid_y_split, valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
````

#### Stochastic Gradient Descent

##### 演算法解釋
在更新參數的時候
gradient descent我們是一次用全部訓練集的數據去計算損失函數的梯度就更新一次參數。
stochastic gradient descent就是一次跑一個樣本或是小批次(mini-batch)樣本然後算出一次梯度或是小批次梯度的平均後就更新一次，那這個樣本或是小批次的樣本是隨機抽取的，所以才會稱為隨機梯度下降法。

##### 輸出準確率 (underfitting
```

    max valid accuracy: 0.7262569832402235

    max valid accuracy: 0.8258426966292135
```

````python
# 模型訓練 (AdaBoost Classifier (ensemble))

from sklearn.model_selection import KFold
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score  

kf = KFold(n_splits=5,             
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x) 

train_acc_list = []
valid_acc_list = []
model_final = None
vacc = 0

for train_index, valid_index in kf.split(train_x):
    train_x_split = train_x.iloc[train_index]
    train_y_split = train_y.iloc[train_index]
    valid_x_split = train_x.iloc[valid_index]
    valid_y_split = train_y.iloc[valid_index]
    
    model = AdaBoostClassifier(n_estimators=500, random_state=0)
    model.fit(train_x_split, train_y_split)
    
    train_pred_y = model.predict(train_x_split)
    train_acc = accuracy_score(train_y_split, train_pred_y)
    valid_pred_y = model.predict(valid_x_split)
    valid_acc = accuracy_score(valid_y_split, valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

    if vacc < valid_acc:
        vacc = valid_acc
        model_final = model
    
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
````

#### AdaBoost Classifier (ensemble)

##### 演算法解釋
AdaBoost方法是一種疊代算法，在每一輪中加入一個新的弱分類器，直到達到某個預定的足夠小的錯誤率。
