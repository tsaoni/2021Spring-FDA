```python
# 載入必要套件

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error as MSE
import numpy as np
import pandas as pd
```

## 載入資料

```python
# 使用 scikit-learn 的資料

dataset = load_boston()
print(type(dataset))
dataset
```

```python
# 觀察資料的 key 值有哪些

dataset.keys()
```

## 資料前處理
### TODO #1: 
- 請將資料以 Pandas 整理成 DataFrame 的型式
- 欄位名稱為 feature_names

```python
df = pd.DataFrame(
    # Please put something here.
    data = dataset.data,
    columns = dataset.feature_names,
)
df['CHAS'] = df['CHAS'].astype(int)
df['RM'] = df['RM'].astype(int)

df
```

```python
print("Number of features: ", len(dataset.feature_names))
```

```python
df['MEDV'] = dataset.target
```

```python
df.head()
```

### TODO #2: 請進行統計各個 feature 的缺失值數目

```python
# Please use some functions of Pandas here.

for col in df.columns:
    print(col, ": ", df[col].isna().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## 觀察資料的特性

### TODO #3: 利用 Seaborn 畫箱型圖，並進行說明
- 觀察每一個 features 的數值分佈
- 可以用來決定是否要對資料的數值進行標準化

```python
f, axes = plt.subplots(4, 3, figsize = (20, 10))
# Plot a boxplot with Seaborn

names = ["CRIM", "ZN", "INDUS", "NOX", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]
for i, t in enumerate(names):
    ax = sns.boxplot(x = "RM",
                 y = t,
                 hue = "CHAS",
                 data = pd.concat([df['RM'], df[t], df['CHAS']], axis = 1),
                 orient = 'v',
                 ax = axes[int(i / 3)][i % 3]
                 )
```

### TODO #4: 利用 Seaborn 進行 Heatmap 作圖，並進行說明
- 觀察每一個特徵與房價 (MEDV) 的關聯程度
- 使用 dr.corr() 來計算 feature 和 feature 之間的關係係數
- 數值愈高表示愈正相關，數值愈低表示愈負相關

```python
plt.figure(figsize=(16, 16))
# Please use a function of Pandas here.
# Plot a heatmap with Seaborn

ax = sns.heatmap(df.corr())
```

### TODO #5: 挑選特徵，並進行說明
- 觀察 heatmap 中具有"顯著"正相關或是"顯著"負相關的特徵
- 把這些特徵進行 concatenation，得到新的 dataframe

```python

X = pd.DataFrame(np.c_[df['ZN'], df['CHAS'], df['RM'], df['DIS'], df['B']], # Please put features inside the bracket
                 columns = ['ZN', 'CHAS', 'RM', 'DIS', 'B']# Please put feature names here 
                )
X
```

### TODO #6: 呈現特徵分佈，並進行說明
- 把你選取到的"重要特徵"進行subplots，畫出散佈圖 (scatter plots)
- 每一張圖的X軸是你選到的特徵、Y軸是房價('MEDV')

```python
# Please plot subplots here
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (16, 7))

sns.scatterplot(data = df, x = 'ZN', y = 'MEDV', hue = 'RM', style = 'CHAS', ax = ax1)
sns.scatterplot(data = df, x = 'B', y = 'MEDV', hue = 'RM', style = 'CHAS', ax = ax2)
sns.scatterplot(data = df, x = 'DIS', y = 'MEDV', hue = 'RM', style = 'CHAS', ax = ax3)
```

## 訓練模型前的準備
### TODO #7: 切分資料
- 利用 train_test_split，將資料切成 train set 和 test set
- 設定 train set 的比例為全部資料的 80%
- 請記得設定 random_state 以控制隨機過程

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, df['MEDV'], test_size = 0.1, random_state = 42)

print(X_train.shape) #(379, 5)
print(X_test.shape) #(127, 5)
print(Y_train.shape) #(379,)
print(Y_test.shape) #(127,)
```

## 訓練模型並進行評估

### TODO #8: 訓練 Linear Regression 模型，並說明訓練結果
- 利用 scikit-learn，將資料 fit 在 LinearRegression
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

```python
# Modeling
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, Y_train)
print("accuracy: ", reg.score(X_train, Y_train))

Y_pred = reg.predict(X_test)
df_y = pd.DataFrame({'pred': Y_pred, 'true': Y_test})
```

```python
# plotting the y_test vs y_pred

sns.scatterplot(data = df_y, x = 'pred', y = 'true')
```

### TODO #9: 訓練 Random Forest Regressor 模型
- 利用 scikit-learn，將資料 fit 在 RandomForestRegressor 模型
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Modeling

regr = RandomForestRegressor(max_depth = 10, random_state = 0)
regr.fit(X_train, Y_train)

print("accuracy: ", regr.score(X_train, Y_train))
Y_pred = regr.predict(X_test)
df_y = pd.DataFrame({'pred': Y_pred, 'true': Y_test})
```

```python
# plotting the y_test vs y_pred

sns.scatterplot(data = df_y, x = 'pred', y = 'true')
```

### TODO #10: 參數調整，並進行說明
- 請嘗試使用 Grid search 來進行調參
    - 可以自己寫 for-loop
    - 或是使用 sklearn.model_selection 中的 GridSearchCV 函式

```python
from sklearn.model_selection import GridSearchCV
```

```python
# for loop

r1 = 20
r2 = 20

for i in range(10, r1):
    for j in range(5, r2):
        regr = RandomForestRegressor(max_depth = i, random_state = j)
        regr.fit(X_train, Y_train)
        print("accuracy: ", regr.score(X_train, Y_train))
```

```python
# Grid Search

s1 = 1
s2 = 1
r1 = 10
r2 = 10
parameters = {'max_depth': np.arange(s1, r1), 'random_state': np.arange(s2, r2)}
regcv = GridSearchCV(regr, parameters)
regcv.fit(X_train, Y_train)

""" for debug
for param in regr.get_params().keys():
    print(param)
"""

best = 0
for i in range(0, (r1-s1) * (r2-s2)):
    for key in sorted(regcv.cv_results_.keys()):
        print(key, ": ", regcv.cv_results_[key][i]) #print all result
    print("\n")
    if regcv.cv_results_['rank_test_score'][i] == 1:
        best = i

print("--------------------------------\n")
for key in sorted(regcv.cv_results_.keys()):
    print(key, ": ", regcv.cv_results_[key][best])
```

### TODO #11: 訓練 其他的 regression 模型，與上述兩種模型比較並說明訓練結果
- 將資料 fit 在其他的迴歸模型，可選擇 SVR、XGBRegressor 或其他...
- 利用 TODO #10 的工具來進行參數調整
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

```python
from sklearn.ensemble import GradientBoostingRegressor

regr = GradientBoostingRegressor(random_state = 0)
regr.fit(X_train, Y_train)

print("accuracy: ", regr.score(X_train, Y_train))
Y_pred = regr.predict(X_test)
df_y = pd.DataFrame({'pred': Y_pred, 'true': Y_test})
```

```python
# Grid Search

s1 = 1
s2 = 1
r1 = 10
r2 = 10
parameters = {'max_depth': np.arange(s1, r1), 'random_state': np.arange(s2, r2)}
regcv = GridSearchCV(regr, parameters)
regcv.fit(X_train, Y_train)

""" for debug
for param in regr.get_params().keys():
    print(param)
"""

best = 0
for i in range(0, (r1-s1) * (r2-s2)):
    for key in sorted(regcv.cv_results_.keys()):
        print(key, ": ", regcv.cv_results_[key][i]) #print all result
    print("\n")
    if regcv.cv_results_['rank_test_score'][i] == 1:
        best = i

print("--------------------------------\n")
for key in sorted(regcv.cv_results_.keys()):
    print(key, ": ", regcv.cv_results_[key][best])
```

```python
# plotting the y_test vs y_pred

sns.scatterplot(data = df_y, x = 'pred', y = 'true')
```
